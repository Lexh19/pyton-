```python

import numpy as np
from scipy.optimize import linprog 

A = np.array([
    [20, 15, 25],  
    [10, 12, 15], 
    [5, 8, 10]    
])


B = np.array([1000, 800, 600])


C = np.array([-1, -1, -1])


result = linprog(C, A_ub=A, b_ub=B)

optimal_units = result.x
optimal_total_units = -result.fun 

print("Optimal units for A, B, C:", optimal_units)
print("Optimal total units:", optimal_total_units)


```

```python
import numpy as np


matrix_A = np.array([[50, 20], [30, 40]])


matrix_B = np.array([[45, 10], [25, 35]])


det_A = np.linalg.det(matrix_A)

det_B = np.linalg.det(matrix_B)

print("Determinant of matrix A:", det_A)
print("Determinant of matrix B:", det_B)
```

Analisis:

Determinan Positif (det > 0): Ini menunjukkan bahwa perusahaan A dan B keduanya memiliki kinerja yang baik. Determinan yang positif biasanya dikaitkan dengan kualitas kinerja yang baik.

Perbandingan Determinan: Det(A) > Det(B) menunjukkan bahwa, berdasarkan matriks ini, perusahaan A memiliki kinerja yang lebih baik daripada perusahaan B. Hal ini dapat diartikan bahwa perusahaan A mungkin lebih efisien dalam penjualan dan produksi bulanannya.

Korelasi dengan Kualitas Kinerja: Meskipun determinan memberikan indikasi umum tentang kualitas kinerja, tetapi tidak memberikan rincian tentang aspek-aspek spesifik dari penjualan dan produksi. Jadi, sementara perusahaan A mungkin memiliki kinerja keseluruhan yang lebih baik, aspek spesifik dari kinerja harus dianalisis lebih lanjut untuk mendapatkan pemahaman yang lebih mendalam.

```python
import numpy as np


A = np.array([[50, 20], [30, 40]])
B = np.array([[45, 10], [25, 35]])


try:
    A_inv = np.linalg.inv(A)
    print("Inverse of matrix A:")
    print(A_inv)
except np.linalg.LinAlgError:
    print("Matrix A is singular, and its inverse does not exist.")


try:
    B_inv = np.linalg.inv(B)
    print("\nInverse of matrix B:")
    print(B_inv)
except np.linalg.LinAlgError:
    print("Matrix B is singular, and its inverse does not exist.")
```
